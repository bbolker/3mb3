# Some examples of using LaTeX in a Jupyter notebook

adapted from [Randall Pruim's examples](http://www.calvin.edu/~rpruim/courses/m343/F12/RStudio/LatexExamples.html)

Jupyter notebooks allows you to mix text, Python code, Python output, Python graphics, and mathematics in a single document. The mathematics is done using a version of $LaTeX$, the premiere mathematics typesetting program. 


Getting in and out of $ \LaTeX $ mode

You enter $ \LaTeX $ mode using `$` (for inline mathematics) or `$$` (for displayed equations). You get back out of $ \LaTeX $ mode using the matching delimiter (either `$` or `$$`.
Some basic mathematics you might use ...

Commands in $ \LaTeX $ are preceded by a backslash.
The simplest commands simply produce a symbol:

| Command  | Result                |
|-----------|----------|
| \log(x)  |	$\log(x) $  |
| \sin(x)  | $ \sin(x) $     |
| A \cup B |	$ A \cup B $ |
| A \cap B |	$ A \cap B $ |
| x \in A  |	$ x \in A $  |
| 5 \pm 2  |   $ 5 \pm 2 $   |



Other commands take additional inputs (placed inside curly braces {}):

| Command |	Result     
|-----------|----------|
| \sqrt{27} 	| $ \sqrt{27} $ |
| \overline{x} 	| $ \overline{x} $ |
| \frac{k}{n} 	| $ \frac{x}{n} $  |
| \binom{k}{n} 	| $ \binom{k}{n} $ |
| \frac{\partial f}{\partial x} |	$ \frac{\partial f}{\partial x} $ |


The underscore (_) and caret (^) have special uses in $\LaTeX$.
(Note the use of `\displaystyle` to get inline mathematics set with larger fonts and more vertical space.)

| Command  |	Result |
|-----------|----------|
| x^2 	| $ x^2 $ |
| x_2 	| $ x_2 $ |
| `\lim_{x\to\infty}`|	$ \lim_{x \to \infty} $ |
| `\displaystyle \lim_{x\to\infty}` |	$ \displaystyle \lim_{x \to \infty} $ |
| `\int_0^{\infty} f(x) \; dx` |	$ \int_0^{\infty} f(x) \; dx $ |
| `\displaystyle \int_0^{\infty} f(x) \; dx` |	$ \displaystyle \int_0^{\infty} f(x) \; dx $ |
