```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
```

```python
npr.seed(101)
x = npr.standard_t(4.5,size=int(1e6))
print((np.mean(x),x.std()))
```

```python
dfvec = np.arange(5,20,0.1)
stdvec =np.array(list(npr.standard_t(df,size=int(1e5)).std() for df in dfvec))
```

```python
%matplotlib inline
plt.plot(dfvec,stdvec,"ro")
plt.plot(dfvec,np.sqrt(dfvec/(dfvec-2)));
```
